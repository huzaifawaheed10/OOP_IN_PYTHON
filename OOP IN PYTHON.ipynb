```python
class Student:

    def __init__(self,name,marks1,marks2,marks3):
        self.name=name
        self.marks1=marks1
        self.marks2=marks2
        self.marks3=marks3
    def calculate_marks(self):
        return (self.marks1+ self.marks2+ self.marks3)/3

    def display(self):
        average=self.calculate_marks()
        print(f"{self.name} Average is {average}")

s1=Student("Malak huzaifa waheed",56,67,78)
s1.display()
```

```python
class Book:

    def __init__(self,title,author,price):
        self.title=title
        self.author=author
        self.price=price

    def display(self):
        print(f"{self.title} is written by {self.author} which has a price of {self.price}$")

s1=Book("Rich Dad Poor Dad","Robert Kiyosaki and Sharon Lechter", 50)
s1.display()

s2=Book("Game changer","Shahid Afridi and Wajahat Saeed Khan",60)
s2.display()
```

```python
class Circle:
    def __init__(self,radius):
        self.radius=radius

    def area(self):
        return 3.14*self.radius**2

    def circumference(self):
        return 2*3.14*self.radius

    def display(self):
        area=self.area()
        circumference=self.circumference()
        print(f"the area is {area} and circumference is {circumference}")

s1=Circle(7)
s1.display()
```

```python
class Student:

    def __init__(self,name,marks):
        self.name=name
        self.marks=marks

    def get_grade(self):
        if self.marks>=80 :
            return "A"
        elif self.marks>=60 :
            return "B"
        elif self.marks>=40 :
            return "C"
        else:
            return "F"

    def display(self):
        print(f"{self.name} grade is {self.get_grade()}")

s1=Student("Mudasir",90)
s1.display()

s2=Student("Hasnain",78)
s2.display()

s3=Student("Baber Azam",39)
s3.display()
```

```python
class BankAccount:

    def __init__(self,name,balance):
        self.name=name
        self.balance=balance
    def deposit(self,amount):
        self.balance+=amount

    def withdraw(self,amount):
         if self.balance>amount:
             self.balance -= amount
         else:
            print("cannot withdraw more than balance")

    def check_balance(self):
         return self.balance

    def display(self):
         print(f"{self.name} balance is {self.balance}")

acc1=BankAccount("Mudasir",90000)
acc1.deposit(10000)
acc1.withdraw(29000)
acc1.display()

```

```python
class Account:
    def __init__(self,balance,account_no):
        self.balance=balance
        self.account_no=account_no

    def debit(self,amount):
        self.balance-=amount
        print(f"Rs {amount} is debited")
        print(f"the balance becomes {self.balance}")
    def credit(self,amount):
        self.balance+=amount
        print(f"Rs {amount} is credited")
        print(f"the balance becomes {self.balance}")
    def show(self):
        print(f"{self.account_no} : balance is Rs {self.balance}")

acc1=Account(120000,1234)
acc1.debit(20000)
acc1.credit(13000)
acc1.show()
```
