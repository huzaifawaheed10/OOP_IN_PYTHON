```python
class Student:

    def __init__(self,name,marks1,marks2,marks3):
        self.name=name
        self.marks1=marks1
        self.marks2=marks2
        self.marks3=marks3
    def calculate_marks(self):
        return (self.marks1+ self.marks2+ self.marks3)/3

    def display(self):
        average=self.calculate_marks()
        print(f"{self.name} Average is {average}")

s1=Student("Malak huzaifa waheed",56,67,78)
s1.display()
```

```python
class Circle:
    def __init__(self,radius):
        self.radius=radius

    def area(self):
        return 3.14*self.radius**2

    def circumference(self):
        return 2*3.14*self.radius

    def display(self):
        area=self.area()
        circumference=self.circumference()
        print(f"the area is {area} and circumference is {circumference}")

s1=Circle(7)
s1.display()
```

```python
class Student:

    def __init__(self,name,marks):
        self.name=name
        self.marks=marks

    def get_grade(self):
        if self.marks>=80 :
            return "A"
        elif self.marks>=60 :
            return "B"
        elif self.marks>=40 :
            return "C"
        else:
            return "F"

    def display(self):
        print(f"{self.name} grade is {self.get_grade()}")

s1=Student("Mudasir",90)
s1.display()

s2=Student("Hasnain",78)
s2.display()

s3=Student("Baber Azam",39)
s3.display()
```

```python
class Account:
    def __init__(self,balance,account_no):
        self.balance=balance
        self.account_no=account_no

    def debit(self,amount):
        self.balance-=amount
        print(f"Rs {amount} is debited")
        print(f"the balance becomes {self.balance}")
    def credit(self,amount):
        self.balance+=amount
        print(f"Rs {amount} is credited")
        print(f"the balance becomes {self.balance}")
    def show(self):
        print(f"{self.account_no} : balance is Rs {self.balance}")

acc1=Account(120000,1234)
acc1.debit(20000)
acc1.credit(13000)
acc1.show()
```

### INHERITANCE

```python
class car:
    @staticmethod
    def start():
        print("Car Started..")
    @staticmethod
    def stop():
        print("Car Stopped..")
class toyota_car(car):

    def __init__(self,name,model):
        self.name=name
        self.model=model

c1=toyota_car("Fortuner",2021)
c2=toyota_car("kia",2025)

print(c1.name,c1.model)
print(c1.start())
```

```python
class Animal:
    @staticmethod
    def eat():
        print("Eating")

class Dog(Animal):
    @staticmethod
    def Bark():
        print("Barking")

print(Dog.eat())
print(Dog.Bark())

```

```python
class Employee:
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary

class Manager(Employee):
    def __init__(self,name,salary,bonus):
        super().__init__(name,salary)
        self.bonus=bonus

    def total_package(self):
        total=self.salary+self.bonus
        print(f"Total  salary of {self.name} is {total} ")

manager1=Manager("Fayaz ullah",65000,14000)
manager1.total_package()

```

### POLYMORPHISM

```python
class complex:
    def __init__(self,real,img):
        self.real=real
        self.img=img

    def show(self):
        print(self.real,"i +",self.img,"j")

    def __add__ (self,num2):                ##dunder function
        newReal=self.real+num2.real
        newImg=self.img+num2.img
        return complex(newReal,newImg)


num1=complex(1,3)
num1.show()

num2=complex(5,7)
num2.show()

num3=num1+num2
num3.show()
```

```python
class complex:
    def __init__(self,real,img):
        self.real=real
        self.img=img

    def show(self):
        print(self.real,"i +",self.img,"j")

    def __sub__ (self,num2):                ##dunder function
        newReal=self.real-num2.real
        newImg=self.img-num2.img
        return complex(newReal,newImg)
num1=complex(1,3)
num1.show()

num2=complex(5,7)
num2.show()

num3=num1-num2
num3.show()
```

```python
class circle:
    def __init__(self,radius):
        self.radius=radius

    def area(self):
        return 3.14*self.radius**2
    def perimeter(self):
        return 3.14*self.radius*2

c1=circle(8)
print(c1.area())
print(c1.perimeter())
```
